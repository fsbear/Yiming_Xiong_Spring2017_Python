```python
import requests

f = open('./API_Key/key.txt', 'r', encoding = 'utf-8', errors = 'ignore')
key = f.read()
payload = {
    'api-key': key,
    'q': "machine learning",
    'page' : 10
}
payload
r = requests.get('https://api.nytimes.com/svc/search/v2/articlesearch.json', params = payload).json()
```
