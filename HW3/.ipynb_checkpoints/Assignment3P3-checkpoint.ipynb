```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/cricket_matches.csv')
df[:2]
```

```python
df1 = df[df['home'] == df['winner']]
df1.head()
```

```python
# df2 = df1[df1['home'] == df1['innings2']]
df1['WinnerScore'] = np.where(df1['home'] == df1['innings2'], df1['innings2_runs'], df1['innings1_runs'])
```

```python
df2 = df1[['home','WinnerScore']]
df2.head()
```

```python
dffinal = pd.DataFrame(df2.groupby(['home'])['WinnerScore'].mean())
dffinal
```

```python
dffinal = dffinal.sort_values(['WinnerScore'], ascending = False)
dffinal.head()
```

```python
dffinal.to_csv('Assignment3P3.csv')
```
