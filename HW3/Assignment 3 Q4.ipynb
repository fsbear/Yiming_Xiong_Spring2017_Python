```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('Data/movies_awards.csv')
df.columns
```

```python
df1 = df.dropna(subset=['Awards'])
df1.head()
```

```python
df2 = df1[['Title', 'Awards']]
df2.head()
```

```python
list1 = df2['Awards'].apply(lambda x:re.findall(r'Won \d+ ([]).',x)[0] if len(re.findall('Won \d+', x))!=0  else 0)
list1
```

```python
df3 = df2.copy()
# df3['Win'] = df3['Awards'].apply(lambda x: re.findall('(\d+) win',x)[0] if(re.findall('\d+ win', x))  else 0)
df3['Golden Globe Nominated'] = df3['Awards'].apply(lambda x: int(re.findall('Nominated for (\d+) Golden Globe',x)[0]) if(re.findall('Nominated for \d+ Golden Globe', x))  else 0)
df3['Golden Globe Won'] = df3['Awards'].apply(lambda x: int(re.findall('Won (\d+) Golden Globe',x)[0]) if(re.findall('Won \d+ Golden Globe', x))  else 0)
df3['Primetime Emmys Nominated'] = df3['Awards'].apply(lambda x: int(re.findall('Nominated for (\d+) Primetime Emmys',x)[0]) if(re.findall('Nominated for \d+ Primetime Emmys', x))  else 0)
df3['Primetime Emmys Won'] = df3['Awards'].apply(lambda x: int(re.findall('Won (\d+) Primetime Emmys',x)[0]) if(re.findall('Won \d+ Primetime Emmys', x))  else 0)
df3['Oscars Nominated'] = df3['Awards'].apply(lambda x: int(re.findall('Nominated for (\d+) Oscars',x)[0]) if(re.findall('Nominated for \d+ Oscars', x))  else 0)
df3['Oscars Won'] = df3['Awards'].apply(lambda x: int(re.findall('Won (\d+) Oscars',x)[0]) if(re.findall('Won \d+ Oscars', x))  else 0)
df3['BAFTA Film Awards Nominated'] = df3['Awards'].apply(lambda x: int(re.findall('Nominated for (\d+) BAFTA Film Awards',x)[0]) if(re.findall('Nominated for \d+ BAFTA Film Awards', x))  else 0)
df3['BAFTA Film Awards Won'] = df3['Awards'].apply(lambda x: int(re.findall('Won (\d+) BAFTA Film Awards',x)[0]) if(re.findall('Won \d+ BAFTA Film Awards', x))  else 0)
```

```python
df3.head()
```

```python
# df3['Win'] = df3['Awards'].apply(lambda x: int(re.findall('(\d+) win',x)[0]) if(re.findall('\d+ win', x))  else 0)
df3['Win'] = df3['Awards'].apply(lambda x: int(re.findall('(\d+) win',x)[0]) if(re.findall('\d+ win', x))  else 0) +df3['Golden Globe Won'] + df3['Primetime Emmys Won'] + df3['Oscars Won'] + df3['BAFTA Film Awards Won']
df3['Nominations'] = df3['Awards'].apply(lambda x: int(re.findall('(\d+) nominations',x)[0]) if(re.findall('\d+ nominations', x))  else 0) + df3['Golden Globe Nominated'] + df3['Primetime Emmys Nominated'] + df3['Oscars Nominated'] + df3['BAFTA Film Awards Nominated']
df3
```

```python
df3.to_csv('Assignment3Q4.csv')
```
