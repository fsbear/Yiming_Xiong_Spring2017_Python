```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/employee_compensation.csv')
df[:1]
```

```python
df1 = df[df['Year Type'] == 'Calendar']
df1[:1]
```

```python
df2 = df1.groupby(['Employee Identifier'])[['Salaries','Overtime', 'Other Salaries', 'Total Salary', 'Retirement', 
                                            'Health/Dental', 'Other Benefits', 'Total Benefits', 
                                            'Total Compensation']].mean()
df2.reset_index(level = 0, inplace = True)
df2.head()
```

```python
df3 = df1[['Employee Identifier', 'Job Family']].drop_duplicates('Employee Identifier').reset_index(drop=True)
```

```python
df3.head()
```

```python
df4 = pd.merge(df3, df2, on = 'Employee Identifier', how = 'left')
df4.head()
```

```python
df4['PercentofOvertime'] = df4['Overtime']/df4['Total Salary']*100
```

```python
df4.head()
```

```python
df5 = df4[df4['PercentofOvertime'] > 5]
df5.head()
```

```python
df6 = df5.groupby(['Job Family'])['Total Benefits', 'Total Compensation'].mean()
df6.head()
```

```python
df6['PercentofCompensation'] = df6['Total Benefits']/df6['Total Compensation']*100
```

```python
df6 = df6.sort_values(['PercentofCompensation'], ascending = False)
df6.head()
```

```python
df6.to_csv('Assignment3Q2P2.csv')
```
