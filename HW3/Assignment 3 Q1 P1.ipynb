```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/vehicle_collisions.csv')
df.head()
```

```python
df['DATE'] = pd.to_datetime(df["DATE"])
df.head()
```

```python
df1 = df[['DATE', 'BOROUGH']]
df1.head()
```

```python
df2 = df1[(df['DATE'] >= '2016-01-01') & (df['DATE'] <= '2016-12-31')]
df2.head()
```

```python
df3 = df2.copy()
df3['MANHATTAN'] = np.where(df3['BOROUGH'] == 'MANHATTAN', 1, 0)
df3.head()
```

```python
 dffinal = pd.groupby(df3,by = df3['DATE'].dt.to_period("M"))
```

```python
dffinal.count()
```

```python
dffinal.sum()
```

```python
dfresult = pd.concat([dffinal.count()['DATE'], dffinal.sum()['MANHATTAN']], axis = 1)
```

```python
dfresult['PERCENTAGE'] = dfresult['MANHATTAN']/dfresult['DATE']
dfresult.columns = ['NYC', 'MANHATTAN', 'PERCENTAGE']
dfresult
```

```python
dfresult.to_csv('Assignment3Q1P1.csv')
```
