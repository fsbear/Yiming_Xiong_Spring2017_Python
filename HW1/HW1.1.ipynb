```python
import nltk
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
words = nltk.corpus.gutenberg.words('austen-persuasion.txt')
print(len(words))
```

```python
wordsdict = {}
a.lower() for a in words:
    if a in wordsdict:
            wordsdict[a] += 1
    else:
        wordsdict[a] = 1
        
print(wordsdict)
```

```python
import operator
sorted_words = sorted(wordsdict.items(), key=operator.itemgetter(1))
sorted_words.reverse()
print (sorted_words[0:9])
```
