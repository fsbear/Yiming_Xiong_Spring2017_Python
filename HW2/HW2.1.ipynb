```python
import nltk

# inaugural_words = [w.lower() for w in nltk.corpus.inaugural.words() if w.isalpha()]
# len(inaugural_words)   132124
import glob

pathList = []
for filepath in glob.glob('/Users/xiongyiming/nltk_data/corpora/inaugural/*.txt'):
    pathList.append(filepath)


```

```python
import string
inaugural_words = []


def readfile():
    for path in pathList:
        f = open(path, 'r', encoding = 'utf-8', errors = 'ignore')
        for line in f:
            for word in line.lower().split():
                word = word.translate(str.maketrans('','',string.punctuation))
                inaugural_words.append(word)

readfile()
```

```python
fdist = nltk.FreqDist(inaugural_words)
fdist.most_common(10)
```

```python
import operator
sorted_words = sorted(fdist.items(), key=lambda fdist: fdist[1], reverse=True)
print (sorted_words[0:9])
```

```python

result = []
i = 0
for a in sorted_words:
    i += 1
    midresult = []
    midresult.append(i)
    midresult.append(a[0])
    midresult.append(a[1])
    
    result.append(midresult)
    
result[:2]
```

```python
import csv
def outPutCsv():
    f = open('/Users/xiongyiming/Documents/INFO7374Python/HW2/HW21.csv', 'a')
    try:
        writer = csv.writer(f)
    


        writer = writer.writerows(result)
    finally:
        f.close()
    
outPutCsv()
print (open('/Users/xiongyiming/Documents/INFO7374Python/HW2/HW21.csv', 'rt').read())
```

```python
%matplotlib inline
from matplotlib import pyplot as plt



x_list = [x for [x, y, z] in result]
z_list = [z for [x, y, z] in result]

plt.loglog(x_list, z_list)
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.title("'Zipf's law")
```
